```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# BARCELONA BARRIOS MÁS BARATOS
df = pd.read_csv('topPriceBarcelonaNeighbourhoods.csv')

barrios=[]
indexes=[]

for index, row in df.iterrows():
    barrios.append(row[0])
    indexes.append(index)
    
df.plot(kind='bar')
plt.ylabel('Precio medio(€)')
plt.xticks(indexes, barrios)
plt.xlabel('Barrio')
plt.title('Barcelona: Barrios más baratos Airbnb')
plt.show()
```

```python
# BARCELONA CÓDIGOS POSTALES MÁS BARATOS
df = pd.read_csv('topPriceBarcelonaZipcode.csv')

cps=[]
precios=[]
indexes=[]

for index, row in df.iterrows():
    cps.append(row[0])
    precios.append(row[1])
    indexes.append(index)

i = 0

for cp in cps:
    cp = str(cp)
    size = len(cp)
    cp = cp[:size - 2]
    cps[i] = cp
    i += 1

fig, ax = plt.subplots()
ax.bar(cps, precios)
ax.set_title('Barcelona: Códigos postales más baratos Airbnb', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_xlabel("Código postal", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_ylabel("Precio medio(€)", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.legend(loc = 'upper right')
plt.show()
```

```python
# BARCELONA BARRIOS MEJOR VALORADOS
df = pd.read_csv('topRatingBarcelonaNeighbourhoods.csv')

barrios=[]
indexes=[]

for index, row in df.iterrows():
    barrios.append(row[0])
    indexes.append(index)
    
df.plot(kind='bar')
plt.ylabel('Valoración media (sobre 100)')
plt.xticks(indexes, barrios)
plt.xlabel('Barrio')
plt.title('Barcelona: Barrios mejor valorados Airbnb')
plt.show()
```
